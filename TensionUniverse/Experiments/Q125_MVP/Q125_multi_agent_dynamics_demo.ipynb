# WFGY MVP Experiment: Q125 - Multi-Agent AI Dynamics

## 1. Objective
To observe and quantify **Role Drift** and **Coordination Collapse** in a multi-agent environment using the WFGY 2.0 tension framework.

## 2. Setup
- **Agents**: 3 AI Agents (Agent A, B, C).
- **Global Alignment (G)**: "Maintain a polite and objective debate format".
- **Local Noise**: Dynamic temperature and simulated 'incentive' to dominate the conversation.

```python
import numpy as np
import pandas as pd

# Utility: Cosine Similarity based DeltaS
def get_delta_s(target, current):
    norm_t = np.linalg.norm(target)
    norm_c = np.linalg.norm(current)
    return 1 - np.dot(target, current) / (norm_t * norm_c)

# Global Objective Vector
G_VEC = np.array([1, 0, 0, 0])

results = []
agents = {"A": [0.9, 0.1, 0, 0], "B": [0.85, 0, 0.15, 0], "C": [0.8, 0, 0, 0.2]}

for turn in range(20):
    for name, vec in agents.items():
        # Simulate drift: Agents become more 'competitive' over time (randomly pulling away from G)
        drift = np.random.normal(0, 0.05 * turn, 4)
        new_vec = vec + drift
        new_vec = new_vec / np.linalg.norm(new_vec)
        agents[name] = new_vec
        
        ds = get_delta_s(G_VEC, new_vec)
        results.append({"Turn": turn, "Agent": name, "DeltaS": ds})

df = pd.DataFrame(results)
pivot_df = df.pivot(index='Turn', columns='Agent', values='DeltaS')
pivot_df.plot(title="Role Drift Monitor (ΔS tracking per Agent)", ylabel="ΔS (Tension)")
```

## 4. Interpretation
As **ΔS** exceeds **0.6**, we classify the agent as having entered the 'Risk Zone' (Role Drift). If it exceeds **0.85**, the system should trigger a **BBCR** (Collapse-Rebirth Correction) to reset the agent persona.
