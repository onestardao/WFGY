# WFGY MVP Experiment: Q107 - Large Scale Collective Action

## 1. Objective
To simulate the tension between individual incentives and collective benefits in large-scale social systems, using **ΔS (Semantic Tension)** to monitor the emergence of "free-riding" behaviors and narrative drift.

## 2. Theoretical Framework (WFGY 2.0)
- **G (Anchor)**: The collective goal (e.g., "Construct a public dam").
- **I (Current state)**: Current narrative/belief field of participants.
- **ΔS**: Measuring the distance between individual actions and collective purpose.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_delta_s(vec_a, vec_b):
    return 1 - np.dot(vec_a, vec_b) / (np.linalg.norm(vec_a) * np.linalg.norm(vec_b))

# Simulation Parameters
NUM_AGENTS = 100
COLLECTIVE_GOAL_VEC = [1.0, 0.0]  # Full alignment

def simulate_turn(agent_beliefs, noise=0.1):
    # Each turn, beliefs drift slightly based on local noise and feedback
    new_beliefs = agent_beliefs + np.random.normal(0, noise, agent_beliefs.shape)
    # Normalize vectors
    new_beliefs = new_beliefs / np.linalg.norm(new_beliefs, axis=1)[:, None]
    return new_beliefs

history_delta_s = []
agent_beliefs = np.random.normal(1.0, 0.2, (NUM_AGENTS, 2))
agent_beliefs = agent_beliefs / np.linalg.norm(agent_beliefs, axis=1)[:, None]

for t in range(50):
    agent_beliefs = simulate_turn(agent_beliefs)
    avg_delta_s = np.mean([calculate_delta_s(b, COLLECTIVE_GOAL_VEC) for b in agent_beliefs])
    history_delta_s.append(avg_delta_s)

print(f"Final Avg ΔS: {history_delta_s[-1]:.4f}")
```

## 3. Visualization
We can see how the 'Tension' increases as agents drift away from the central goal.

```python
plt.plot(history_delta_s)
plt.title("Semantic Tension (ΔS) Over Time (Collective Action Drift)")
plt.xlabel("Turn")
plt.ylabel("Avg ΔS")
plt.show()
```
